```python
def agregar_registros():
    # Inicializar el diccionario
    datos = {
        "Nombre": [],
        "Subtítulo": [],
        "Etiqueta1": [],
        "Etiqueta2": [],
        "Color1": [],
        "Color2": [],
        "Coste Maná": [],
        "Coste Reciclaje": [],
        "Color Reciclaje": [],
        "Might": [],
        "Tipo": [],
        "Rareza": [],
        "Keyword1": [],
        "Keyword2": [],
    }

    while True:
        #print("\nPor favor, introduce los datos para cada campo:")
        
        for clave in datos.keys():
            # Pedir input para cada campo y almacenar la respuesta
            valor = input(f"{clave}: ")
            datos[clave].append(valor)

        # Preguntar si quiere agregar otro registro
        agregar_otro = input("\n¿Quieres añadir otro registro? (si/no): ").strip().lower()
        if agregar_otro != "si":
            break

    print("\nDatos finales:")
    print(datos)

    return datos
```
